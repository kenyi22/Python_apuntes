```python
def es_primo(numero):
  if numero==2:
    return True
  elif numero % 2 ==0:
    return False
  else:
    for divisor in range(3,numero,2):
      if numero % divisor ==0:
        return False
    return True
```

```python
es_primo(6)
```

```python
#usando listas
def son_primos(lista_numeros):
  lista_primos=[]

  for numero in lista_numeros:
    lista_primos.append(es_primo(numero))
  
  return lista_primos
```

```python
#ejercicio: Pangrams

import string

string.ascii_lowercase
```

```python
def es_pangram(frase):
  alfabeto=string.ascii_lowercase
  frase=frase.lower()
  caracteres=list(set(frase))

  for caracter in alfabeto:
    if caracter not in caracteres:
      return False
  
  return True
```

```python
es_pangram("Hola me llamo manuel")
```

```python
es_pangram("The quick brown fox jumps over")
```

#ejercicio para determinar si un numero es perfecto

```python
def divisores_propios(numero):
  return [divisor for divisor in range(1,numero) if numero % divisor ==0]
```

```python
divisores_propios(6)
```

```python
def es_perfecto(numero):
  if numero == sum(divisores_propios(numero)):
    return True
  else:
    return False
```

```python
es_perfecto(15)
```
